# **Splitting**

```python
!pip install --quiet langchain langchain_community pypdf
```

```python
from langchain.document_loaders import WebBaseLoader
loader = WebBaseLoader("https://nlp-weekly-update.hashnode.dev/weekly-deep-learning-focus-mastering-the-essentials")
```

```python
docs = loader.load()
```

```python
len(docs)
```

```python
docs
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter,CharacterTextSplitter
```

```python
chunk_size = 150
chunk_overlap = 50
r_text_splitter = RecursiveCharacterTextSplitter(chunk_size=chunk_size, chunk_overlap=chunk_overlap)
c_text_splitter = CharacterTextSplitter(chunk_size=chunk_size, chunk_overlap=chunk_overlap)
```

```python
docs
```

```python
if docs:
  for doc in docs:
    chunks = r_text_splitter.split_text(doc.page_content)

len(chunks)
```

```python
chunks
```

# **Semantic Splittters**

```python
!pip install --quiet langchain_experimental
```

```python
from langchain_experimental.text_splitter import SemanticChunker
```

```python
!pip install --quiet langchain_experimental langchain_openai
```

```python
text_splitter = SemanticChunker(
    embed_model , breakpoint_threshold_type='percentile'
)
```

```python
semantic_splits = text_splitter.split_text(docs[0].page_content)
```

```python
len(semantic_splits)
```

```python
for splits in semantic_splits:
  print(splits)
  print('----------------------------------------------')
```
